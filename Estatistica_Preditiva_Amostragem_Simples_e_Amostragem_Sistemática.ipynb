# Amostragem Sistemática

```python
def amostragem_sistematica(df, step):

  indexes = np.arange(3, len(df), step=step)
  amostra_sistematica = df.iloc[indexes]
  return amostra_sistematica

amostra_sistematica = amostragem_sistematica(vendas, 3)
amostra_sistematica.head()
```

# Comparando o describe da amostra_sistematica X vendas.

```python
amostra_sistematica.describe()
```

```python
vendas.describe()
```

### Podemos perceber que também a amostragem sistemática permanece muito próxima da população, o que indica que a amostragem também fiel a população
